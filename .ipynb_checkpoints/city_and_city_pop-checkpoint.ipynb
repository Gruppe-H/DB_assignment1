```python
import pandas as pd
```

```python
df1 = pd.read_csv('DBassign1/2016-Cities.csv')
df2 = pd.read_csv('DBassign1/2016-GHG_Emissions.csv')
df3 = pd.read_csv('DBassign1/2017-Community_Emissions.csv')
df4 = pd.read_csv('DBassign1/2017-Cities_Emission_Reduction_Targets.csv')
df5 = pd.read_csv('DBassign1/2023-Cities_Climate_Risk.csv')

dataframes = [df1, df2, df3, df4, df5]
```

```python
# Columns in Cities table 
c_id = 'city_id'
c_name = 'city_name'
c_loc = 'city_location'
c_country = 'country_id'
c_alt = 'average_altitude'
c_temp = 'average_temp'

# Columns in City_Populations table
cp_id = 'population_id'
cp_amount = 'population_amount'
cp_year = 'population_year'

# placeholder for country
co_name = 'country_name'
# placeholder for city short name
c_sname = 'city_short_name'
# placeholder other types of city name
c_lname = 'city_placeholder_name'
```

```python
for df in dataframes:
    df.rename(columns={'Country': co_name, 'Country/Area': co_name,
                       'City Short Name':c_sname, 'City Name':c_lname, 'City':c_lname, 'City Location':c_loc,'Average altitude (m)':c_alt, '​Average altitude (m)':c_alt,
                       'Average annual temperature (in Celsius)​':c_temp,
                       'Current Population':cp_amount, 'Population': cp_amount, 'Population Year':cp_year}, inplace=True)
```

```python
for df in dataframes:
    print(df.columns)
```

```python
#for Cities table
combined_df = pd.concat([
    df1[[co_name, c_sname]],
    df2[[co_name, c_lname, c_sname, c_loc, c_alt, c_temp]],
    df3[[co_name, c_lname, c_loc, c_alt, c_temp]],
    df4[[co_name, c_lname, c_loc]],
    df5[[co_name, c_lname, c_loc]],
])

newdf = combined_df.drop_duplicates().reset_index(drop=True)
newdf.nunique()
```

```python
newdf[c_name] = None

for index, row in newdf.iterrows():
    if not pd.isna(row[c_lname]) and not pd.isna(row[c_sname]):
        if row[c_lname] == row[c_sname]:
            newdf.at[index, c_name] = row[c_lname]
        else:
            newdf.at[index, c_name] = row[c_lname]
    elif not pd.isna(row[c_lname]):
        newdf.at[index, c_name] = row[c_lname]
    elif not pd.isna(row[c_sname]):
        newdf.at[index, c_name] = row[c_sname]

# Drop unnecessary columns
#df.drop([c_lname, c_sname], axis=1, inplace=True)

# Drop duplicate rows based on 'c_name'
#df = df.drop_duplicates(subset=['c_name'], keep='first').reset_index(drop=True)

newdf
```

```python
city_name_counts = newdf[c_name].value_counts()
print(city_name_counts)
```

```python
test_df = newdf[newdf[c_name] == 'Canberra']
test_df
```

```python
import numpy as np

merged_rows = []  # List to store merged rows
for name, group in newdf.groupby(c_name):
    merged_row = {col: group[col].dropna().iloc[0] if group[col].notna().any() else np.nan for col in newdf.columns}
    merged_rows.append(merged_row)

merged_df = pd.DataFrame(merged_rows)

merged_df
```

```python
test_df = merged_df[merged_df[c_name] == 'Canberra']
test_df
```

```python
merged_df.drop([c_lname, c_sname], axis=1, inplace=True)
merged_df
```

```python
country_df = pd.read_csv('DBassign1/Table files/csv/country_data.csv')
country_df = country_df.sort_values(by='country_name', ascending=True)
country_df.dropna(how='all')
country_df.reset_index(drop=True, inplace=True)
country_df['country_id'] = country_df.index + 1 

country_df
```

```python
city_df = merged_df

c_c_merged_df = pd.merge(city_df, country_df, on='country_name', how='left')

city_df['country_id'] = c_c_merged_df['country_id']

city_df
```

```python
city_df.dropna(how='all')
city_df.reset_index(drop=True, inplace=True)
city_df['city_id'] = city_df.index + 1 
city_df.drop([co_name], axis=1, inplace=True)

city_df
```

```python
city_df.to_csv('cities_table.csv')
```

```python
# City Populations table
# df1 has no population data at all
com_cp_df = pd.concat([
    df2[[c_lname,cp_amount]],
    df3[[c_lname,cp_amount]],
    df4[[c_lname,cp_amount, cp_year]],
    df5[[c_lname,cp_amount, cp_year]],
])

com_cp_df[c_name] = com_cp_df[c_lname]
com_cp_df.drop([c_lname], axis=1, inplace=True)
com_cp_df
```

```python
city_df = pd.read_csv('cities_table.csv')
city_df = city_df[['city_name', 'city_id']]
city_df
```

```python
c_mapping = city_df.set_index('city_name')['city_id']
com_cp_df['city_id'] = com_cp_df['city_name'].map(c_mapping)
com_cp_df.drop(columns=['city_name'], inplace=True)

com_cp_df
```

```python
com_cp_df.reset_index(drop=True, inplace=True)
com_cp_df[cp_id] = com_cp_df.index + 1 

com_cp_df
```
