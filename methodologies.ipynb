```python
import pandas as pd
```

```python
# Only one with methodology
df1 = pd.read_csv('DBassign1/2016-GHG_Emissions.csv')
```

```python
# Columns in Methodologies table
m_id = 'methodology_id'
m = 'primary_methodology'
m_det = 'details'
m_eid = 'emission_id'

# placeholders to find correct m_eid
org_no = 'org_no'
```

```python
df1.rename(columns={'Primary Methodology': m, 'Methodology Details': m_det, 'Account Number': org_no}, inplace=True)
```

```python
tmp_m_df = df1[[m, m_det,org_no]]
tmp_m_df
```

```python
tmp_m_df.reset_index(drop=True, inplace=True)
tmp_m_df[m_id] = tmp_m_df.index + 1
tmp_m_df
```

```python
e_df = pd.read_csv('DBassign1/emission_table.csv', sep=';')
e_df
```

```python
emission_mapping = e_df[['Organization_id', 'emission_id']].set_index('Organization_id')['emission_id']

tmp_m_df[m_eid] = tmp_m_df[org_no].map(emission_mapping)

tmp_m_df.drop(columns=[org_no], inplace=True)

tmp_m_df
```

```python
tmp_m_df.to_csv('DBassign1/Table files/csv/methodologies_table.csv')
```
