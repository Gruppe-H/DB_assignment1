```python
import pandas as pd
```

```python
df1 = pd.read_csv('DBassign1/2016-Cities.csv')
df2 = pd.read_csv('DBassign1/2016-GHG_Emissions.csv')
df3 = pd.read_csv('DBassign1/2017-Community_Emissions.csv')
df4 = pd.read_csv('DBassign1/2017-Cities_Emission_Reduction_Targets.csv')
df5 = pd.read_csv('DBassign1/2023-Cities_Climate_Risk.csv')
dataframes = [df1, df2, df3, df4, df5]
```

```python
# Columns in Comments table
c_id = 'comment_id'
c = 'comment'
c_year = 'comment_year'
c_eid = 'emission_id'

# placeholders to find correct c_eid
c_gas = 'gasses_incl'
org_no = 'org_no'
```

```python
for df in dataframes:
    df.rename(columns={'Comment': c, 'Organization Number': org_no,'Account No': org_no,'Account Number': org_no,'Account number': org_no,'Gases included': c_gas}, inplace=True)
```

```python
# it is only df1, -3 and -4 that has comments
df1[c_year] = '2016'
df3[c_year] = '2017'
df4[c_year] = '2017'
```

```python
#for Cities table
combined_df = pd.concat([
    df1[[c, c_year, org_no]],
    df3[[c, c_year, c_gas, org_no]],
    df4[[c, c_year, org_no]],
])

c_df = combined_df.drop_duplicates().reset_index(drop=True)
c_df.nunique()
```

```python
c_df
```

```python
c_df.dropna(subset=[c], inplace=True)
c_df
```

```python
e_df = pd.read_csv('DBassign1/emission_table.csv', sep=';')
e_df
```

```python
emission_mapping = e_df[['Organization_id', 'emission_id', 'gases_included']].set_index('Organization_id')['emission_id']

c_df['emission_id'] = c_df[org_no].map(emission_mapping)

c_df.drop(columns=[org_no, c_gas], inplace=True)

c_df
```

```python
c_df.reset_index(drop=True, inplace=True)
c_df['comment_id'] = c_df.index + 1 
c_df
```

```python
def write_sql_insert_file(file_name, df, table_name):    
    insert_queries = []
    with open(file_name, 'w', encoding='utf-8') as f:
        for index, row in df.iterrows():
            insert_query = f"INSERT INTO {table_name} ("
            for column_name in df.columns:
                insert_query += f" {column_name},"
            # Remove the trailing comma
            insert_query = insert_query[:-1] + ") VALUES ("
            for column_name in df.columns:
                value = row[column_name]
                if isinstance(value, str):
                    # Check if the value contains a single quote, and if so, escape it by adding another single quote
                    value = value.replace("'", "''")
                else:
                    # If the value is not a string, leave it as is
                    pass
                insert_query += f" '{value}',"
            # Remove the trailing comma and complete the query
            insert_query = insert_query[:-1] + ");\n"
            insert_query = insert_query.replace("'nan'", "NULL")
            f.write(insert_query)
```
