```python
import pandas as pd
```

```python
# this csv file is the only one with scopes
df1 = pd.read_csv('DBassign1/2017-Community_Emissions.csv')
```

```python
# Columns in Scopes table
s_id = 'scopes_id'
s_name = 'scope_name'

# Columns in Scope_Emissions table
se_id = 'scope_emission_id'
s_id
s_em = 'scope_emission'
s_eid = 'emission_id'

# placeholders
org_no = 'org_no'
s_incl = 'scope_incl'
s_em1 = 'emission_1'
s_em2 = 'emission_2'
```

```python
# Scopes Included ,Total Scope 1 Emissions (metric tonnes CO2e),Total Scope 2 Emissions (metric tonnes CO2e),
df1.rename(columns={'Scopes Included ': s_incl, 'Total Scope 1 Emissions (metric tonnes CO2e)': s_em1, 'Total Scope 2 Emissions (metric tonnes CO2e)':s_em2,'Account number': org_no}, inplace=True)
```

```python
com_s_df = df1[[s_incl,s_em1,s_em2,org_no]]
com_s_df
```

```python
#Scopes table
scope_data = {
    s_id: [1, 2, 3],
    s_name: ['Total Scope 1', 'Total Scope 2', 'Waste Scope 3 (Total BASIC emissions)']
}

s_df = pd.DataFrame(scope_data)
#s_df.to_csv('DBassign1/Table files/csv/scopes_table.csv')
```

```python
# temp id
com_s_df.reset_index(drop=True, inplace=True)
com_s_df[se_id] = com_s_df.index + 1

# Scope_Emission table
se_df = pd.melt(com_s_df, id_vars=[se_id, org_no], value_vars=['emission_1', 'emission_2'], var_name=s_id, value_name=s_em)
se_df[s_id] = se_df[s_id].str.split('_').str[-1].astype(int)
#se_df.drop(columns=['id'], inplace=True)

se_df.dropna(subset=[s_em], inplace=True)

se_df
```

```python
e_df = pd.read_csv('DBassign1/emission_table.csv', sep=';')
e_df
```

```python
emission_mapping = e_df[['Organization_id', 'emission_id']].set_index('Organization_id')['emission_id']

se_df[s_eid] = se_df[org_no].map(emission_mapping)

se_df.drop(columns=[org_no], inplace=True)

se_df
```
