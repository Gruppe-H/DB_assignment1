```python
import pandas as pd

df1 = pd.read_csv('DBassign1/2016-Cities.csv')
df2 = pd.read_csv('DBassign1/2016-GHG_Emissions.csv')
df3 = pd.read_csv('DBassign1/2017-Community_Emissions.csv')
df4 = pd.read_csv('DBassign1/2017-Cities_Emission_Reduction_Targets.csv')
df5 = pd.read_csv('DBassign1/2023-Cities_Climate_Risk.csv')

dataframes = [df1, df2, df3, df4, df5]
```

```python
# variables for column names
region_name = 'region_name'
cdp_region = 'CDP_region'
country_name = 'country_name'
```

```python
# column names for Countries
#co_name = 'country_name'
co_loc = 'country_location'
```

```python
for df in dataframes:
    df.rename(columns={'Country': country_name,'Region':region_name, 'Country/Area':country_name,'CDP Region':cdp_region, 'Country Location':co_loc}, inplace=True)
```

```python
#for region table
combined_df = pd.concat([
    df1[[country_name]],
    df2[[country_name]],
    df3[[region_name,country_name]],
    df4[[region_name,country_name]],
    df5[[cdp_region,country_name]],
])

# Then all duplicates are dropped to get unique pairs
newdf = combined_df.drop_duplicates().reset_index(drop=True)
newdf.nunique()
```

```python
newdf
```

```python
newdf[cdp_region].unique()
```

```python
newdf[region_name].unique()
```

```python
newdf.sort_values(by='region_name', ascending=False, inplace=True)

# Drop duplicates based on city_name, keeping the first occurrence
cleaned_df = newdf.drop_duplicates(subset='country_name', keep='first')
cleaned_df
```

```python
cleaned_df.nunique()
```

```python
cleaned_df.to_csv('cleaned_data.csv', index=False)
```

```python
#for Countries table
combined_country_df = pd.concat([
    df1[[country_name, co_loc]],
    df2[[country_name,co_loc]],
    df3[[country_name,co_loc]],
    df4[[country_name,co_loc]],
    df5[[country_name]],
])

# Then all duplicates are dropped to get unique pairs
new_country_df = combined_country_df.drop_duplicates().reset_index(drop=True)
new_country_df.nunique()
```

```python
new_country_df
```

```python
new_country_df.sort_values(by='country_location', ascending=False, inplace=True)

# Drop duplicates based on city_name, keeping the first occurrence
cleaned_country_df = new_country_df.drop_duplicates(subset='country_name', keep='first')
cleaned_country_df
```

```python
country_with_FK = pd.merge(cleaned_df, cleaned_country_df, on='country_name', how='outer')
country_with_FK
```

```python
merged_df = country_with_FK
merged_df.loc[merged_df['region_name'] == 'South and West Asia', 'region_name'] = 1
merged_df.loc[merged_df['region_name'] == 'South Asia and Oceania', 'region_name'] = 2
merged_df.loc[merged_df['region_name'] == 'North America', 'region_name'] = 3
merged_df.loc[merged_df['region_name'] == 'Latin America', 'region_name'] = 4
merged_df.loc[merged_df['region_name'] == 'Europe', 'region_name'] = 5
merged_df.loc[merged_df['region_name'] == 'East Asia', 'region_name'] = 6
merged_df.loc[merged_df['region_name'] == 'Africa', 'region_name'] = 7
merged_df
```

```python
merged_df.loc[merged_df['CDP_region'] == 'Latin America', 'CDP_region'] = 1
merged_df.loc[merged_df['CDP_region'] == 'East Asia', 'CDP_region'] = 2
merged_df.loc[merged_df['CDP_region'] == 'Europe', 'CDP_region'] = 3
merged_df.loc[merged_df['CDP_region'] == 'United States of America', 'CDP_region'] = 4
merged_df.loc[merged_df['CDP_region'] == 'Southeast Asia', 'CDP_region'] = 5
merged_df.loc[merged_df['CDP_region'] == 'Oceania', 'CDP_region'] = 6
merged_df.loc[merged_df['CDP_region'] == 'Africa', 'CDP_region'] = 7
merged_df.loc[merged_df['CDP_region'] == 'South Asia', 'CDP_region'] = 8
merged_df.loc[merged_df['CDP_region'] == 'Canada', 'CDP_region'] = 9
merged_df.loc[merged_df['CDP_region'] == 'Middle East', 'CDP_region'] = 10
merged_df
```
